```python
import numpy as np
import matplotlib.pyplot as plt

#  
n_simulations = 1000  #   
sample_sizes = [10, 50, 100, 500, 1000]  #    
a = 0  #     
b = 1  #    
true_mean = (a + b) / 2  #    = (a + b)/2
true_var = (b - a)**2 / 12  #   = (b-a)^2/12

#      n_simulations 
#       
for sample_size in sample_sizes:
    #      
    means = [np.random.uniform(a, b, sample_size).mean() 
             for _ in range(n_simulations)]
    
    #     :
    # empirical_mean -     
    # empirical_std -    
    empirical_mean = np.mean(means)  #     
    empirical_std = np.std(means)    #    
    # 
    plt.figure(figsize=(10, 6))
    plt.hist(means, bins=30, density=True, alpha=0.7,
             label=f' \n={empirical_mean:.3f}\n={empirical_std:.3f}')
    plt.axvline(true_mean, color='r', linestyle='--', 
                label=f' ={true_mean}')
    
    plt.title(f'  \n  n={sample_size}')
    plt.xlabel('')
    plt.ylabel('')
    plt.legend()
    plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
#    x   
# lim (2x^2 - 3x - 5)/(x + 1)  x+

#           x (x^2)
# lim (2 - 3/x - 5/x^2)/(1/x + 1/x^2) = 2

#      
def f(x):
    return (2*x**2 - 3*x - 5)/(x + 1)

#    x
x = np.linspace(1, 100, 1000)

#   
y = f(x)

#  
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='f(x)')
plt.axhline(y=2, color='r', linestyle='--', label='y = 2')
plt.title('  f(x) = (2x^2 - 3x - 5)/(x + 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

print(f"   x+  2")




```
